```python
import numpy as np
import pandas as pd
import sympy as sym
from matplotlib import pyplot as plt
```

```python
def derivative(x):
    return 2*(x+3)
def function(x):
    return (x+3)**2
```

```python
x_coord=np.linspace(-15,15,100)
plt.plot(x_coord,function(x_coord))
plt.plot(2,function(2),"rs")
```

```python
def gradientDescent(learningRate,startValue,maxIterations):
    xCoordList=list()
    xCoordList.append(startValue)
    x=startValue
    for i in range (maxIterations):
        slope=derivative(x)
        x=x-(learningRate*slope)
        xCoordList.append(x)
    print("Local Minima occurs at:",x)
    return xCoordList
```

```python
x_coord=np.linspace(-15,15,100)
plt.plot(x_coord,function(x_coord))

points=gradientDescent(0.05,2,100)

xs=np.array(points)
plt.plot(xs,function(xs),"r.-")
```
